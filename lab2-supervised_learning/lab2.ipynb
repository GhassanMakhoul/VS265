<h2 align="center">1. Linear Neuron with sigmoidal output nonlinearity </h2> 

Derive the modified learning rule for a linear neuron with sigmoidal output nonlinearity:

$$y= \sigma(u) = \frac{1}{1 + e^{-u}}$$
with $u = w^T x = \sum_i w_i x_i$

$$ Error = \Sigma_i \hspace{2mm} |T_i - y_i|^2 $$
The first simplification willl be to eliminate the summation over all the weights. Since we are taking the partial derivative with respect to one weight, all the other elements of the summation will be treated as constants, thus "go to zero" anyway.  Thus we can drop the sum. After this I expanded the partial derivative using the chain rule. 
$$ \frac{\partial E}{\partial w_i} = \frac{\partial E} {\partial y} \frac{\partial y}{\partial u} \frac{\partial u}{\partial w_i} \implies 2| T_i - \delta(w^Tx)|\delta(w^Tx)(1-(\delta(w^Tx))x_i$$
